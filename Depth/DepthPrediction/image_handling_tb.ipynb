```python
image_segs, depth_segs, masks = load_dataset_segments('nyu_depth_v2_labeled.mat',
                                                 images=(1030,),
                                                 no_superpixels=400,
                                                 x_window_size=84,
                                                 y_window_size=84)
edges = find_neighbors(np.squeeze(mask))
gc_array = graph_cut_pairwise_array(image_segs, edges, distance_type=1)
```

```python
no_sections = len(image_segs)
lbp_similarities = np.zeros((no_sections,))
colorhist_similarities = np.zeros((no_sections,))
color_similarities = np.zeros((no_sections,))

n = 100
for ii in range(0, no_sections):
    color_similarities[ii] = logistic_color_diff(image_segs[n, ...], image_segs[ii, ...], gamma=4e-5)
    colorhist_similarities[ii] = logistic_color_hist_diff(image_segs[n, ...], image_segs[ii, ...], gamma=1e-4)
    lbp_similarities[ii] = logistic_lbp_diff(image_segs[n, ...], image_segs[ii, ...], gamma=1e-3, points=4, radius=2)
```

```python
fig14, axes14 = plt.subplots(1, 1, figsize=(8,6), dpi=800)
axes14.plot(range(0, no_sections), color_similarities, 'b', label='Color Distance')
axes14.plot(range(0, no_sections), colorhist_similarities, 'r', label='Color Histogram Distance')
axes14.plot(range(0, no_sections), lbp_similarities, 'k', label='Local Binary Pattern Distance')
axes14.legend()
axes14.set_xlabel('Image Patch Number')
axes14.set_ylabel('Logistically Regressed Distance Value')
fig14.savefig('distance_metric_output.pdf')
```

```python
difference_100 = np.zeros((0,))
difference_200 = np.zeros((0,))
difference_400 = np.zeros((0,))
for n in range(999, 1020):
    [image_segs, depth_segs, masks] = load_dataset_segments('nyu_depth_v2_labeled.mat',
                                                            images=(n,),
                                                            no_superpixels=100,
                                                            x_window_size=83,
                                                            y_window_size=83)
    depths_sp = apply_depths(depth_segs, np.squeeze(masks))
    difference_100 = np.hstack((difference_100,np.array(depths[n, ...]).ravel() - depths_sp.ravel()))
    
    [image_segs, depth_segs, masks] = load_dataset_segments('nyu_depth_v2_labeled.mat',
                                                            images=(n,),
                                                            no_superpixels=200,
                                                            x_window_size=83,
                                                            y_window_size=83)
    depths_sp = apply_depths(depth_segs, np.squeeze(masks))
    difference_200 = np.hstack((difference_200,np.array(depths[n, ...]).ravel() - depths_sp.ravel()))
    
    [image_segs, depth_segs, masks] = load_dataset_segments('nyu_depth_v2_labeled.mat',
                                                            images=(n,),
                                                            no_superpixels=400,
                                                            x_window_size=83,
                                                            y_window_size=83)
    depths_sp = apply_depths(depth_segs, np.squeeze(masks))
    difference_400 = np.hstack((difference_400,np.array(depths[n, ...]).ravel() - depths_sp.ravel()))
```

```python
cdf_x_points = np.arange(0.0,0.5,0.01)
cdf_val_100 = np.zeros(cdf_x_points.shape)
cdf_val_200 = np.zeros(cdf_x_points.shape)
cdf_val_400 = np.zeros(cdf_x_points.shape)
for val_idx in range(0, len(cdf_x_points)):
    cdf_val_100[val_idx] = np.sum(np.abs(difference_100) < cdf_x_points[val_idx]) / float(len(difference_100))
    cdf_val_200[val_idx] = np.sum(np.abs(difference_200) < cdf_x_points[val_idx]) / float(len(difference_200))
    cdf_val_400[val_idx] = np.sum(np.abs(difference_400) < cdf_x_points[val_idx]) / float(len(difference_400))
```

```python
counts, edges = np.histogram(difference_400, bins=100, normed=True)
centers = (edges[1:] + edges[0:-1]) / 2

fig19, axes19 = plt.subplots(1, 1, figsize=(8,6), dpi=800)
axes19.bar(centers, counts, width=0.1)
axes19.set_xlim([-0.5, 0.5])
axes19.set_ylabel('Estimated PDF value')
axes19.set_xlabel('Distance Difference (m)')
fig19.savefig('superpixel_constant_distance_assumption.pdf')
```

```python
fig20, axes20 = plt.subplots(1, 1, figsize=(5,4), dpi=800)
axes20.plot(cdf_x_points, cdf_val_100, 'r-.', label='100 Superpixels')
axes20.plot(cdf_x_points, cdf_val_200, 'b--', label='200 Superpixels')
axes20.plot(cdf_x_points, cdf_val_400, 'k-', label='400 Superpixels')
axes20.set_ylabel('Estimated CDF value')
axes20.set_xlabel('Distance Difference Magnitude (m)')
axes20.legend(loc=4)
fig20.savefig('superpixel_constant_distance_assumption_cdf_multiple.pdf')
```
