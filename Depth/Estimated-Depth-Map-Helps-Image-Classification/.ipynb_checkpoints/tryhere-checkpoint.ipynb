```python
# coding: utf-8

# In[59]:
import sys
from cifar2im import load_CIFAR_batch
from neural_net import *
import matplotlib.pyplot as plt
import time

dataset_dir='data_batch'
start_time=time.time()
```

```python
def get_CIFAR10_data(num_training=49000, num_validation=1000, num_test=1000):
    X, y = load_CIFAR_batch (dataset_dir)
    X_val = X[-num_validation-num_test:-num_test]
    y_val = y[-num_validation-num_test:-num_test]
    X_train = X[:-num_validation-num_test]
    y_train = y[:-num_validation-num_test]
    X_test = X[-num_test:]
    y_test = y[-num_test:]
    mean_image = np.mean(X_train, axis=0)
    X_train -= mean_image
    X_val -= mean_image
    X_test -= mean_image
    return X_train, y_train, X_val, y_val, X_test, y_test


# Invoke the above function to get our data.
X_train, y_train, X_val, y_val, X_test, y_test = get_CIFAR10_data()
print "finish loading"
print 'Train data : ', X_train.shape
print 'Validation data : ', X_val.shape
print 'Test data: ', X_test.shape
print "Time", (time.time()-start_time)/60.0
```

```python
from neural_net import *

input_size = X_train.shape[1]
hidden_size = 150
num_classes = 10

net = TwoLayerNet(input_size, hidden_size, num_classes,1e-4)
stats = net.train(X_train, y_train, X_val, y_val,
                            num_iters=70000, batch_size=128,
                            learning_rate=1e-5, learning_rate_decay=0.99,
                            reg=0, verbose=True,update="momentum",arg=0.9,dropout=0.5)
```

```python
# In[126]:

val_acc = (net.predict(X_train) == y_train).mean()
print 'Train accuracy: ', val_acc
val_acc = (net.predict(X_val) == y_val).mean()
print 'Validation accuracy: ', val_acc

val_acc = (net.predict(X_test) == y_test).mean()
print 'Test accuracy: ', val_acc

print "time",(time.time()-start_time)/60.0 

# In[121]:
```
